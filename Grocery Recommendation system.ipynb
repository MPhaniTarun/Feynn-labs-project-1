```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df=pd.read_csv("C:\\Users\\TARUN\\Downloads\\Groceries_dataset.csv\\Groceries_dataset.csv")
```

```python
df
```

```python
all_unique_products = df['itemDescription'].unique()
print("Total products: {}".format(len(all_unique_products)))
```

```python
one_hot_encoding = pd.get_dummies(df['itemDescription'])
df.drop('itemDescription', inplace=True, axis=1)
df = df.join(one_hot_encoding)
```

```python
df.head(5)
```

```python
# Grouping the transactions based on Member_number and Date

grpby_transactions = df.groupby(["Member_number","Date"])[all_unique_products[:]].apply(sum)
grpby_transactions = grpby_transactions.reset_index()[all_unique_products]
```

```python
grpby_transactions.head(5)
```

```python
def one_hot_conversion(data):
    for item in all_unique_products:
        if data[item] > 0:
            data[item] = item
    return data

grpby_transactions = grpby_transactions.apply(one_hot_conversion, axis=1)
grpby_transactions.head(5)
```

```python
x = grpby_transactions.values
x = [sub[~(sub == 0)].tolist() for sub in x if sub[sub != 0].tolist()]
transactions = x
```

```python
transactions
```

```python
rules = apriori(transactions,min_support=0.0003,min_confidance=0.05,min_lift=3,min_length=2,target="rules")
association_rules_pred = list(rules)
```

```python
i=0
for item in association_rules_pred:
    pair = item[0] 
    items = [x for x in pair]
    print( items[0] + " -> " + items[1])
```
